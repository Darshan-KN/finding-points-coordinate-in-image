```python

#other modify code part of above one

import cv2
import numpy as np
import pandas as pd
import os

# Function to detect and number points in an image
def detect_and_number_points(image_path, output_excel_folder, annotated_image_folder):
    # Load the image
    img = cv2.imread(image_path)
    
    # Convert image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Threshold the image to get a binary image
    _, binary = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)
    
    # Find contours in the binary image
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    points = []
    point_numbers = []

    # Iterate over the contours to find and number points
    for i, contour in enumerate(contours):
        # Calculate the centroid of each contour
        M = cv2.moments(contour)
        if M['m00'] != 0:
            cx = int(M['m10'] / M['m00'])
            cy = int(M['m01'] / M['m00'])
            points.append((cx, cy))
            point_numbers.append(i + 1)
            
            # Draw the contour and number the points
            cv2.drawContours(img, [contour], -1, (0, 0, 0), 1)
            cv2.putText(img, str(i + 1), (cx, cy), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)

    # Create filenames based on the image name
    filename = os.path.splitext(os.path.basename(image_path))[0]
    
    # Save the annotated image
    annotated_image_path = os.path.join(annotated_image_folder, f'{filename}_annotated.png')
    cv2.imwrite(annotated_image_path, img)

    # Save the coordinates and point numbers to an Excel file
    output_excel_path = os.path.join(output_excel_folder, f'{filename}_points.xlsx')
    df = pd.DataFrame(points, columns=['X', 'Y'])
    df['Point Number'] = point_numbers
    df.to_excel(output_excel_path, index=False)

# Example usage
image_path = r"C:\\Users\\admin\\Desktop\\y_gen_96.png"  # Replace with your image file path
output_excel_folder = r"C:\\Users\\admin\\Desktop\\New folder (8)"  # Replace with your output Excel folder path
annotated_image_folder = r"C:\\Users\\admin\\Desktop\\New folder (8)"  # Replace with your annotated image folder path

detect_and_number_points(image_path, output_excel_folder, annotated_image_folder)
```
